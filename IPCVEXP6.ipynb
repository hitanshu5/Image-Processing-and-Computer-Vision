```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def add_gaussian_noise(image, mean=0, std_dev=20):
    gauss = np.random.normal(mean, std_dev, image.shape)
    noisy_image = cv.add(image.astype(int), gauss.astype(int))
    return noisy_image

def add_salt_and_pepper_noise(image):
    row, column = image.shape

    num_pixels_white = np.random.randint(300, 8000)
    num_pixels_black = np.random.randint(300, 8000)

    new_image = np.array(image)

    for _ in range(num_pixels_white):
        y_coord = np.random.randint(0, row - 1)
        x_coord = np.random.randint(0, column - 1)
        new_image[y_coord, x_coord] = 255

    for _ in range(num_pixels_black):
        y_coord = np.random.randint(0, row - 1)
        x_coord = np.random.randint(0, column - 1)
        new_image[y_coord, x_coord] = 0

    return new_image

def median_filter(image, kernel_size=(3, 3)):
    height, width = image.shape
    k_h, k_w = kernel_size

    filtered_image = np.zeros((height, width), dtype=image.dtype)

    for i in range(height - k_h + 1):
        for j in range(width - k_w + 1):
            patch = image[i:i + k_h, j:j + k_w]
            median_value = np.median(patch)
            filtered_image[i + k_h // 2, j + k_w // 2] = median_value

    return filtered_image

# Load an image
image = cv.imread('/content/bird.jpeg', cv.IMREAD_GRAYSCALE)

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Add Gaussian noise
image_with_gaussian_noise = add_gaussian_noise(image)

# Add salt and pepper noise
image_with_salt_and_pepper_noise = add_salt_and_pepper_noise(image)

# Apply median filter to remove Gaussian noise
filtered_image_gaussian = median_filter(image_with_gaussian_noise)

# Apply median filter to remove salt and pepper noise
filtered_image_salt_and_pepper = median_filter(image_with_salt_and_pepper_noise)

# Display the images
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_with_gaussian_noise, cmap='gray')
plt.title('Before Filtering (Gaussian)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image_gaussian, cmap='gray')
plt.title('After Filtering (Gaussian)')
plt.axis('off')

plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def add_gaussian_noise_rgb(image, mean=0, std_dev=20):
    gauss = np.random.normal(mean, std_dev, image.shape)
    noisy_image = cv.add(image.astype(int), gauss.astype(int))
    return noisy_image

def add_salt_and_pepper_noise_rgb(image):
    row, column, channel = image.shape
    num_pixels_white = np.random.randint(300, 8000)
    num_pixels_black = np.random.randint(300, 8000)
    new_image = np.array(image)

    for _ in range(num_pixels_white):
        y_coord = np.random.randint(0, row - 1)
        x_coord = np.random.randint(0, column - 1)
        new_image[y_coord, x_coord] = [255] * channel

    for _ in range(num_pixels_black):
        y_coord = np.random.randint(0, row - 1)
        x_coord = np.random.randint(0, column - 1)
        new_image[y_coord, x_coord] = [0] * channel

    return new_image

def median_filter_rgb(image, kernel_size=(3, 3)):
    filtered_image = np.zeros_like(image)

    for c in range(image.shape[2]):
        filtered_image[:, :, c] = median_filter(image[:, :, c], kernel_size=kernel_size)

    return filtered_image

# Load an RGB image
grayscale_image = cv.imread('/content/image.jpeg')

# Add Gaussian noise to the RGB image
image_with_gaussian_noise = add_gaussian_noise_rgb(grayscale_image)

# Add salt and pepper noise to the RGB image
image_with_salt_and_pepper_noise = add_salt_and_pepper_noise_rgb(grayscale_image)

# Apply median filter to remove Gaussian noise from the RGB image
filtered_image_gaussian = median_filter_rgb(image_with_gaussian_noise)

# Apply median filter to remove salt and pepper noise from the RGB image
filtered_image_salt_and_pepper = median_filter_rgb(image_with_salt_and_pepper_noise)

# Display the images
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(grayscale_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_with_gaussian_noise)
plt.title('Before Filtering (Gaussian)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_image_gaussian)
plt.title('After Filtering (Gaussian)')
plt.axis('off')

plt.show()
```
